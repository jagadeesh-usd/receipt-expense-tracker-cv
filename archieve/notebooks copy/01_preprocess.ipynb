```python
from pathlib import Path
import sys
sys.path.append("../")
from src.preprocess import save_preprocessed
```

```python
RAW_DIR = Path(f"{gpath}/data/raw/SROIE2019")
PROCESSED_DIR = Path(f"{gpath}/data/processed/SROIE2019")
```

```python
image_folders = ["train", "test"]

for split in image_folders:
    img_dir = RAW_DIR / split / "img"
    dest_dir = PROCESSED_DIR / split

    print(f"Processing: {img_dir} → {dest_dir}")

    for img_path in img_dir.glob("*"):
        dest = dest_dir / img_path.name
        save_preprocessed(str(img_path), str(dest))

print("✔ Preprocessing complete!")
```
